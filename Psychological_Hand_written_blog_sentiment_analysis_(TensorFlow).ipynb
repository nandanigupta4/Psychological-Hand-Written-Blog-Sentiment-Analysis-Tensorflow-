```python
from transformers import AutoTokenizer

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

```python
raw_inputs = [
    """To Self-Justify is Human
Self-Serving Bias refers to a tendency for individuals to make dispositional attributions for their success and situational attributions for their failure. For example, in a social situation like an interview, if an interviewee named Reema nails the interview, chances are that she will attribute it to her intrinsic knowledge and negotiation skills. On the other hand, if she fails to pass it, she might claim that interviewer asked too difficult questions that anyone would have failed to answer or interviewer didn't seem to like her or in the worst case, that luck was not on her side. Researchers have gathered a great deal of evidence in support of the informal observation that we take credit for the good and deny the bad. This is not a lone case. Say for example you were travelling by a public bus and upon reaching the destination you realized that your phone is missing, you are likely to claim that someone has stolen it or you lost it because the bus was heavily crowded, instead of addressing your carelessness.
During the first outbreak of Covid-19, a mass exodus of migrant labourers occurred. One of the main reasons behind it was the fear that people had in their minds. It is also attributed as one of the chief reasons for the rife of the virus across states. There were indeed genuine concerns involved in this situation, but spread of fake news was also a contributing factor. Spreading of fake news is generally associated with people having vested interests. But this does not unveils the whole picture. People also spread fake news to justify their fears!
My goal here is not to analyse this situation from a political or economical stand-point. It is just to provide a relatable example for the readers to understand, "Why we as humans, tend to justify our actions, beliefs, and feelings?" from a social psychological stand-point and present the observations researchers made on the basis of data and surveys.
In the aftermath of an earthquake in India, investigators collected and analysed the rumours being spread. What they found was rather startling: when the disaster occurred in a neighborhoring village such that the residents in question could feel the tremors but were not in imminent danger, there was an abundance of rumours forecasting impending impending doom. The residents not only believed but also, helped spread rumours to the effect, that 1) a flood was rushing towards them; 2) Feb 26 would be a day of deluge and destruction; 3) there would another earthquake on the lunar eclipse and so on.
The kind of rumours spread in the case of earthquake case sound strikingly similar to what happened during the migrant labourer mass exodus outbreak, don't they? Perhaps, their cause of their spread also has some correlation!
Why on Earth would anyone in his right senses invent, believe, and communicate such stories? Were these people paranoids? Were these people masochists? Certainly these rumours would not encourage the people to feel calm and secured. One rather compelling explanation is that the people were terribly scared, and because there was not ample justification for this fear, they invented their own justification. Hence, they would not have to feel foolish in the eyes of the world. This explanation is bolstered by yet another study of rumours wherein it turned out that rumours spread in an Indian village directly affected by a disaster of similar magnitude was much less. People were scared, they had good reasons to be frightened; they had no need to seek additional justification for their fears.
Leon Festinger organized this array of findings and used them as the basis for a powerful theory of human motivation, the theory of cognitive dissonance. Cognitive dissonance is a state of tension that occurs whenever an individual simultaneously holds two cognitions ( ideas, attitudes, beliefs, opinions) that are psychologically consistent. Stated differently, two cognitions are dissonant if, when considered alone, the opposite of one follows from other. Because the occurrence of cognitive dissonance is unpleasant, people are motivated to reduce it. To hold two ideas that contradict with each other is to flirt with absurdity, and- as Albert Camus, the existentialist philosopher, has observed- humans are creatures who spend their lives trying to convince that existence is not absurd.
How do we convince ourselves that our lives are not absurd- that is, how do we reduce cognitive dissonance? By changing one or both cognitions in such a way to render them more compatible ( more consonant) with each other or by adding mote cognitions that help bridge the gap between the original cognitions.
Let's understand this with the help of a familiar example. Suppose a person consumes a lot of sugar and then reads a report of the medical evidence linking sugar consumption to diabetes. The person experiences dissonance. The cognition "I eat sugar" is dissonant with the cognition "excessive sugar causes diabetes." Clearly, the most efficient way for this person to reduce dissonance in such a situation is to give up sugar. The cognition "excessive sugar causes diabetes" is consonant with the cognition "I don't eat sugar. "
But, quitting sugar might not be very easy for people. What will they do reduce dissonance then? In all probability they will try to work on the other cognition.
A similar study was made about smoking in 1964, the evidence of smoking causes lung cancer was published. The survey indicated that 90% of the non smokers in the study justified and believed in the evidence ( their cognition was in consonance with the evidence). Smoking is a difficult habit to break; only 9% of the smokers had been able to quit. To justify the continuity of activity, smokers tried to debunk the report. They were likely to deny the evidence; 40% of the heavy smokers said a link had not been proven to exist. Over twice as many smokers as non smokers agreed that there are many hazards in life and both smokers and non smokers get cancer ( Again, attempts to reduce social dissonance).
(The concepts talked of here are backed by the findings from the book, Social Animal by Elliot Aronson.)"""
]
inputs = tokenizer(raw_inputs, padding=True, truncation=True, return_tensors="tf")
print(inputs)
```

```python
from transformers import TFAutoModel

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
model = TFAutoModel.from_pretrained(checkpoint)
```

```python
from transformers import TFAutoModel  ## this class does not output probabilities

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
model = TFAutoModel.from_pretrained(checkpoint)
```

```python
outputs = model(inputs)
print(outputs.last_hidden_state.shape)
```

```python
from transformers import TFAutoModelForSequenceClassification ## transformer library has modules like this for each nlp task

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
model = TFAutoModelForSequenceClassification.from_pretrained(checkpoint)
outputs = model(inputs)
```

```python
print(outputs.logits.shape)
```

```python
print(outputs.logits)
```

```python
import tensorflow as tf

predictions = tf.math.softmax(outputs.logits, axis=-1)
print(predictions)
```
